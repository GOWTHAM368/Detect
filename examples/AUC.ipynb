```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_r0AE = pd.read_csv("../tests/auc_RISH0_AutoEncoder_clean.csv", header=None)
df_r0PCA = pd.read_csv("../tests/auc_RISH0_PCA_CNV85.csv", header=None)
df_r0Z = pd.read_csv("../tests/auc_RISH0_Z-score_clean.csv", header=None)
```

### CNV

```python
x_R0 = [df_r0AE.values.mean(), df_r0PCA.values.mean(), df_r0Z.values.mean()] 
std_R0 = [df_r0AE.values.std(), df_r0PCA.values.std(), df_r0Z.values.std()]
x_labels = ['AE', 'PCA', 'Z']
colors_R0 = ['#c54630', '#c54630', '#c54630']
edgecol = "black"
linspace = [np.zeros(len(df_r0AE.values)), np.ones(len(df_r0AE.values)), 2*np.ones(len(df_r0AE.values))]
swarm_r0 = [df_r0AE.values, df_r0PCA.values, df_r0Z.values]
```

```python
import matplotlib.cm as cm
import seaborn as sns
sns.set_style("white")
from matplotlib.colors import Normalize
my_cmap = cm.get_cmap('Blues')
my_norm = Normalize(vmin=0, vmax=6)
fig, ax = plt.subplots(1,1,figsize=(8, 8))
#ax.set_ylim(ylimError)

ax.set_xlim(-1,3)
ax.set_ylabel("AUC",size=28)
ax.tick_params(axis='y',labelsize=24)
ax.tick_params(axis='x',labelsize=24)
ax.set_xticks(np.arange(0, 3, 1))
ax.set_yticks(np.arange(0.3, 1.01, 0.1))
ax.set_ylim(0.3,1.1)
ax.set_title("R0", size=36)
ax.set_xticklabels(x_labels, rotation=0)
ax.spines['left'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.hlines(0.5,-1,9,linestyles='dashed',linewidth=3)

ax.bar(np.arange(0,3),x_R0, color=colors_R0, alpha=1,  linewidth=2, edgecolor=edgecol, width=1, yerr=std_R0, capsize=8, label="R0", error_kw=dict(lw=2, capsize=8, capthick=2, ecolor=edgecol))

# R0 statistical annotation
x1, x2 = 0, 2   # columns 'Sat' and 'Sun' (first column: 0, see plt.xticks())
y1 = np.max(x_R0)+np.max(std_R0)+0.1
y2 = np.max(x_R0)+np.max(std_R0)+0.1
ax.plot([x1, x2],[y1,y2], lw=1.5,c="black")
ax.text((x1+x2)*.5, y1, "*", ha='center', va='bottom', color="black", fontsize=28)

x1, x2 = 1, 2   # columns 'Sat' and 'Sun' (first column: 0, see plt.xticks())
y1 = np.max(x_R0)+np.max(std_R0) 
y2 = np.max(x_R0)+np.max(std_R0)
ax.plot([x1, x2],[y1,y2], lw=1.5,c="black")
ax.text((x1+x2)*.5, y1, "*", ha='center', va='bottom', color="black", fontsize=28)

ax.scatter(linspace,swarm_r0, zorder=10, alpha=0.2, color="black")

sns.despine()

#plt.legend(fontsize=24)

fig.tight_layout()
fig.savefig('AUC_CNV_r0.svg', dpi=200)
```
